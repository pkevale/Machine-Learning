```python
def get_origin(manufacturer):
    manufacturer = str(manufacturer).lower()
    
    if manufacturer in ['harley-davidson', 'chevrolet', 'pontiac', 'ram', 'ford', 'gmc', 'tesla', 
                        'jeep', 'dodge', 'cadillac', 'chrysler', 'lincoln', 'buick', 'saturn', 'mercury']:
        return 'American'
    
    elif manufacturer in ['lexus', 'nissan', 'toyota', 'acura', 'honda', 'infiniti', 'subaru', 
                          'mitsubishi', 'datsun', 'mazda']:
        return 'Japanese'
    
    elif manufacturer in ['volkswagen', 'mercedes-benz', 'bmw', 'audi', 'porsche']:
        return 'German'
    
    elif manufacturer in ['ferrari', 'fiat', 'alfa-romeo']:
        return 'Italian'
    
    elif manufacturer in ['kia', 'hyundai']:
        return 'Korean'
    
    elif manufacturer in ['volvo']:
        return 'Swedish'
    
    elif manufacturer in ['rover', 'mini', 'land rover', 'jaguar']:
        return 'English'
    
    else:
        return 'Other'

df['manufacturer_origin'] = df['manufacturer'].apply(get_origin)

```

```python
def get_vehicle_type(vehicle_type):
    vehicle_type = str(vehicle_type).lower()
    
    if vehicle_type in ['sedan', 'convertible', 'coupe', 'hatchback', 'other']:
        return 'luxury_small'
    
    elif vehicle_type in ['suv', 'wagon']:
        return 'luxury_large'
    
    elif vehicle_type in ['pickup', 'truck', 'offroad']:
        return 'non-luxury_small'
    
    elif vehicle_type in ['van', 'mini-van', 'bus']:
        return 'non-luxury_large'
    
    else:
        return 'Other'

df['type_category'] = df['type'].apply(get_vehicle_type)
```

```python
def classify_paint_color(color):
    color = str(color).lower() if pd.notnull(color) else 'unknown'
    
    if color in ['white', 'silver', 'yellow', 'orange', 'green', 'custom']:
        return 'Light Color'
    elif color in ['black', 'red', 'blue', 'purple', 'grey', 'brown']:
        return 'Dark Color'
    else:
        return 'Other'

df['paint_category'] = df['paint_color'].apply(classify_paint_color)

```

```python
# Step 1: Convert all values to string before extracting numbers
df['cylinders'] = df['cylinders'].astype(str).str.extract(r'(\d+)')

# Step 2: Convert to numeric (optional: for modeling)
df['cylinders'] = pd.to_numeric(df['cylinders'], errors='coerce')
```

```python
x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=123456)
```
